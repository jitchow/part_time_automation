```python
from langdetect import detect

categories = {
    '剧情韩漫': 1,
    '耽美漫画': 3,
    '真人漫画': 4,
    '校园漫画': 5,
    '日韩漫画': 6,
    '少女漫画': 8,
    '恋爱漫画': 9
}

def select_category(title, desc):
    try:
        if any(keyword in title or keyword in desc for keyword in ['学', '校', '教']):
            return categories['校园漫画']
        elif any(keyword in title or keyword in desc for keyword in ['女']):
            return categories['少女漫画']
        elif any(keyword in title or keyword in desc for keyword in ['爱']):
            return categories['恋爱漫画']
        elif detect(title) == 'ja' or detect(desc) == 'ja':
            return categories['日韩漫画']
        else:
            return categories['剧情韩漫']
    except Exception as e:
        return categories['剧情韩漫'] 
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
import time

# Step 1: Open the link
url = "https://madmin.9xyrp3kg4b86.com/index/index"
service = Service(executable_path='msedgedriver.exe')
driver = webdriver.Edge(service=service)
driver.get(url)

driver.maximize_window()

# Step 2: Input 'xiaoyu' to the username textbox
username_input = driver.find_element(By.ID, "username")
username_input.send_keys("xiaoyu")

# Step 3: Input 'qweqwe' to the password textbox
password_input = driver.find_element(By.ID, "password")
password_input.send_keys("qweqwe")

# Step 4: Click the login button
login_button = driver.find_element(By.XPATH, "//input[@value='登录']")
login_button.click()

# Wait for the page to load after login
time.sleep(2)

# Click on the '漫画管理' button using a different XPath
manga_management_button = driver.find_element(By.XPATH, "//a[contains(., '漫画管理')]")
manga_management_button.click()

time.sleep(1)

# Click on the '漫画列表' button
manga_list_button = driver.find_element(By.XPATH, "//a[contains(., '漫画列表')]")
manga_list_button.click()

time.sleep(2)

iframe_tab_id = "c92761fd166a5c85385340d8c1a3f456"
driver.switch_to.frame(driver.find_element(By.CSS_SELECTOR, f"iframe[tab-id='{iframe_tab_id}']"))

# Click on the visible input element to trigger the dropdown
status_input = driver.find_element(By.XPATH, "//input[@placeholder='选择状态']")
status_input.click()
time.sleep(1)
status_option = driver.find_element(By.XPATH, "//dd[@lay-value='0']")
status_option.click()

search_button = driver.find_element(By.XPATH, "//input[@value='搜索']")
search_button.click()

time.sleep(2)

# Step 8: Select '100 条/页' from the list
items_per_page_dropdown = driver.find_element(By.XPATH, "//span[@class='layui-laypage-limits']//select")
items_per_page_dropdown.click()
items_per_page_option = driver.find_element(By.XPATH, "//option[@value='100']")
items_per_page_option.click()

time.sleep(2)

# Step 9: Go to the last page (120 in this case)
last_page_button = driver.find_element(By.XPATH, "//a[@class='layui-laypage-last']")
last_page_button.click()

for i in range(100):
    # Locate the first row of the table
    first_row = driver.find_element(By.XPATH, "//tbody/tr[1]")

    # Locate and click the '编辑' button in the first row
    edit_button = first_row.find_element(By.XPATH, "//button[contains(@onclick, 'set_edit')]")
    edit_button.click()

    # Locate the iframe element
    iframe = driver.find_element(By.ID, "layui-layer-iframe1")
    # Switch to the iframe
    driver.switch_to.frame(iframe)

    # Extract value from the input element
    title_input = driver.find_element(By.NAME, "title")
    title_value = title_input.get_attribute("value") if title_input else ''

    # Extract value from the textarea element
    desc_textarea = driver.find_element(By.ID, "desc")
    desc_value = desc_textarea.get_attribute("value") if desc_textarea else ''

    category_index = select_category(title=title_value, desc=desc_value)

    # Click on the visible input element to trigger the dropdown
    category_input = driver.find_element(By.XPATH, "//input[@placeholder='选择题材']")
    category_input.click()
    time.sleep(1)
    status_option = driver.find_element(By.XPATH, f"//dd[@lay-value='{category_index}']")
    status_option.click()

    # Wait for the radio button to be clickable
    normal_radio = driver.find_element(By.XPATH, "//div[text()='正常']")
    normal_radio.click()

    # click save
    save_button = driver.find_element(By.CSS_SELECTOR, "div.layui-input-block button")
    save_button.click()

    time.sleep(2)

    driver.switch_to.default_content()
    iframe_tab_id = "c92761fd166a5c85385340d8c1a3f456"
    driver.switch_to.frame(driver.find_element(By.CSS_SELECTOR, f"iframe[tab-id='{iframe_tab_id}']"))

    print(i+1)

# Close the browser
driver.quit()
```
